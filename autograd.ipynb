AutoGrad


```python
x= torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
w = torch.tensor(1.,requires_grad=True)
b = torch.tensor(0.,requires_grad=True)
```

```python
z = w*x+b
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
loss = F.binary_cross_entropy(y_pred, y)
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
